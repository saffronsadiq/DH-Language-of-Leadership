```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from nltk.corpus import stopwords
import matplotlib.pyplot as plt

# Load your data (assuming it's a pandas DataFrame)
df = pd.read_csv("trump_speeches.csv")  # Change to your actual data path

# List of emotion labels
emotion_labels = ['optimistic', 'angry', 'fearful', 'proud', 'empathetic', 'determined', 'critical']

# Function to filter speeches based on emotion and generate n-gram frequencies
def get_frequent_phrases(df, emotion_label):
    # Filter rows where the emotion label is True
    filtered_speeches = df[df[emotion_label] == True]['RawText']
    
    # Tokenize the speeches and remove stop words (adjusting for n-grams)
    stop_words = stopwords.words('english')
    vectorizer = CountVectorizer(stop_words=stop_words, ngram_range=(2, 6), max_features=1000)  # Capture 2- to 6-word n-grams
    X = vectorizer.fit_transform(filtered_speeches)
    
    # Get n-gram counts and convert to a DataFrame
    ngram_count = X.sum(axis=0).A1
    ngrams = vectorizer.get_feature_names_out()
    ngram_freq = pd.DataFrame({'phrase': ngrams, 'count': ngram_count})
    ngram_freq = ngram_freq.sort_values(by='count', ascending=False)
    
    return ngram_freq

# Plotting the most frequent n-grams for each emotion
def plot_ngram_frequency(ngram_freq, emotion_label):
    top_phrases = ngram_freq.head(20)  # Top 20 n-grams
    plt.figure(figsize=(10, 6))
    plt.barh(top_phrases['phrase'], top_phrases['count'], color='skyblue')
    plt.xlabel('Phrase Count')
    plt.title(f"Top 20 Most Frequent Phrases for '{emotion_label}' Emotion")
    plt.gca().invert_yaxis()  # Reverse the y-axis to show the most frequent phrases at the top
    plt.show()

# Loop through each emotion and generate frequent phrases
for emotion in emotion_labels:
    ngram_freq = get_frequent_phrases(df, emotion)
    plot_ngram_frequency(ngram_freq, emotion)
```

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from nltk.corpus import stopwords
import matplotlib.pyplot as plt

# Load your data (assuming it's a pandas DataFrame)
df = pd.read_csv("trump_speeches.csv")  # Change to your actual data path

# List of emotion labels
emotion_labels = ['optimistic', 'angry', 'fearful', 'proud', 'empathetic', 'determined', 'critical']

# Function to filter speeches based on emotion and generate n-gram frequencies
def get_frequent_phrases(df, emotion_label):
    # Filter rows where the emotion label is True
    filtered_speeches = df[df[emotion_label] == True]['RawText']
    
    # Check if the filtered speeches have data
    if filtered_speeches.empty:
        print(f"No data for {emotion_label}")
        return pd.DataFrame()  # Return empty DataFrame if no data
    
    # Tokenize the speeches and remove stop words (adjusting for n-grams)
    stop_words = stopwords.words('english')  # Use NLTK stopwords
    vectorizer = CountVectorizer(stop_words=stop_words, ngram_range=(2, 6), max_features=1000)  # Capture 2- to 6-word n-grams
    X = vectorizer.fit_transform(filtered_speeches)
    
    # Get n-gram counts and convert to a DataFrame
    ngram_count = X.sum(axis=0).A1
    ngrams = vectorizer.get_feature_names_out()
    ngram_freq = pd.DataFrame({'phrase': ngrams, 'count': ngram_count})
    ngram_freq = ngram_freq.sort_values(by='count', ascending=False)
    
    return ngram_freq

# Plotting the 20th to 40th most frequent n-grams for each emotion
def plot_ngram_frequency(ngram_freq, emotion_label):
    if ngram_freq.empty:
        return  # Skip plotting if no n-grams were found
    
    # Get 20th to 40th most frequent n-grams
    ngram_range = ngram_freq.iloc[19:50]  # 20th to 40th (remember indexing starts at 0)
    plt.figure(figsize=(10, 6))
    plt.barh(ngram_range['phrase'], ngram_range['count'], color='skyblue')
    plt.xlabel('Phrase Count')
    plt.title(f"20th to 40th Most Frequent Phrases for '{emotion_label}' Emotion")
    plt.gca().invert_yaxis()  # Reverse the y-axis to show the most frequent phrases at the top
    plt.show()

# Loop through each emotion and generate frequent phrases
for emotion in emotion_labels:
    print(f"Processing emotion: {emotion}")  # Print which emotion is being processed
    ngram_freq = get_frequent_phrases(df, emotion)
    plot_ngram_frequency(ngram_freq, emotion)
```
