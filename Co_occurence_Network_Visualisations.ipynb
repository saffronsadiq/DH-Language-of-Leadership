```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt

# Load the CSV file
csv_file_path = "trump_co-occurence_keywords.csv"  # Replace with your CSV file path
df = pd.read_csv(csv_file_path)

# Define themes and corresponding colors
themes = {
    "Crisis Terms": ["emergency", "threat", "disaster", "broken system", "danger", "urgency", "peril", "instability", "at risk", "system failure", "terror"],
    "Optimistic Terms": ["progress", "prosperity", "innovation", "future", "hope", "advancement", "vision", "American", "national", "pride", "proud", "constitution", "constitutional", "liberty", "America", "justice", "freedom", "values"],
    "Equality Terms": ["inequality", "equity", "integrity", "equality", "compassion", "civil liberties", "civil rights"],
    "Economic Terms": ["jobs", "cost", "taxes", "cost-effective", "rising costs", "inflation", "wealth", "financial", "business", "unemployment", "tax", "economy", "economic", "green new deal", "climate", "emission"],
    "Healthcare and Social Terms": ["health", "health care"],
    "Border and Security Terms": ["security", "borders", "immigration", "asylum", "deportation", "undocumented", "fencing", "detention center", "border", "military", "criminal", "arms"],
    "Political Institutions": ["supreme court", "senate", "bill"]
}

theme_colors = {
    "Crisis Terms": "red",
    "Optimistic Terms": "green",
    "Equality Terms": "yellow",
    "Economic Terms": "orange",
    "Healthcare and Social Terms": "purple",
    "Border and Security Terms": "brown",
    "Political Institutions": "pink"
}

# Initialize a graph
G = nx.Graph()

# Create positions for themes
theme_positions = {}
offset = 0
for i, theme in enumerate(themes.keys()):
    theme_positions[theme] = (offset, 0)
    offset += 5  # Spacing between clusters

# Add edges and nodes
for _, row in df.iterrows():
    category = row['Category']  # Theme keyword
    word = row['Word']          # Co-occurrence word
    frequency = row['Frequency']  # Frequency of co-occurrence

    # Add nodes and edges only if the category is part of themes
    for theme, keywords in themes.items():
        if category in keywords:
            G.add_node(theme, pos=theme_positions[theme], color=theme_colors[theme], size=700)  # Theme node
            G.add_node(word, color=theme_colors[theme], size=frequency * 20)  # Co-occurrence word node
            G.add_edge(theme, word, weight=frequency)

# Extract positions for layout
pos = nx.spring_layout(G, k=1.5, pos={node: data['pos'] for node, data in G.nodes(data=True) if 'pos' in data}, fixed=theme_positions.keys())
node_colors = [data['color'] for _, data in G.nodes(data=True)]
node_sizes = [data['size'] for _, data in G.nodes(data=True)]
edge_weights = [G[u][v]['weight'] * 0.1 for u, v in G.edges()]  # Scale edge weight for visualization

# Draw the graph
plt.figure(figsize=(15, 15))

# Draw nodes
nx.draw_networkx_nodes(G, pos, node_size=node_sizes, node_color=node_colors, alpha=0.9)

# Draw edges
nx.draw_networkx_edges(G, pos, width=edge_weights, edge_color="gray", alpha=0.7)

# Draw labels for themes and co-occurrence words
nx.draw_networkx_labels(G, pos, font_size=10, font_color="black")

# Title and show
plt.title("Thematic Co-occurrence Network with Color-Coded Nodes", fontsize=16)
plt.axis("off")
plt.show()
```
