```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = "trump_speeches.csv"  # Replace with "biden_speeches.csv" for Biden's speeches
df = pd.read_csv(file_path)

# Ensure your dataset has the correct emotion columns
emotions = ['optimistic', 'angry', 'fearful', 'proud', 'empathetic', 'determined', 'critical']

# Calculate the distribution
emotion_counts = df[emotions].sum()

# Plot the distribution
plt.figure(figsize=(10, 6))
emotion_counts.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title("Emotion Distribution", fontsize=16)
plt.ylabel("Count", fontsize=14)
plt.xlabel("Emotion", fontsize=14)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
